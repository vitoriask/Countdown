```python
#Finalizado
#tempos invlidos

import time
global hf
global mf
global sf
inicio = time.localtime()
a = inicio.tm_hour
b = inicio.tm_min
c = inicio.tm_sec

print(a,b,c)

h = int(input("Horas: "))
m = int(input("Minutos: "))
s = int(input("Segundos: "))

sf = (60 - c + s)
#s
if sf >  60:
  sf = sf - 60

else:
  print("w")

#m

if m < b:
  ma = m - 1
  m1 = (60 - b + ma) 
  mf = m1 - 60

elif m > b:
  m1 = b + 1
  mf = m - m1

else:
  print("w")


#hora

if h > a:
  h1 = h - 1
  hf = h - a

elif h < a:
  print("The hour is invalid!")

elif h == a:
  hf = 0

else:
  print("w")

contador = 0
totals = sf + (mf*60) + (hf*60*60)
tot = sf + (mf*60) + (hf*60*60)
tempo = sf + (mf*60) + (hf*60*60)


while contador != totals:
  tot = tot - 1
  print(contador)
  print("Faltam: ",tot,"segundos.")
  contador += 1
  time.sleep(0)
if tot == 0:
  print("Contagem encerrada!")
else:
  print("Contagem encerrada!")




print("Descrio: "

)

print('Horrio de incio da contagem: ',a,b,c)
print("Horrio de encerramento da contagem: ",h,m,s)
print('Tempo gasto: ',hf,mf,sf)
print(totals)


print(h,m,s)
print(a,b,c,)

print(hf,mf,sf)
```
